```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
# root = os.path.abspath(os.curdir)
# print(root)
```

```python
url_2020 = 'https://en.anpec.org.br/previous-editions/2020/artigos-2020.html'
r = requests.get(url_2020)
    
# Parsing the html page
soup = BeautifulSoup(r.text, 'html.parser')
print(soup.prettify())
```

```python
areas = [i.a.get_text().strip() for i in  
         soup.find_all('h4', {'class': "panel-title"})]
print(areas)
```

```python
lista = []

for i in range(1,11):
    lista.append(i**2)

print(lista)
lista2 = [i**2 for i in range(1,11)]
print(lista2)
```

```python
# Getting links and title of papers
tags_a = [i.a for i in soup.find_all('li')]
print(tags_a)
```

```python
try:
    links = [x['href'] for x in tags_a]
except TypeError:
    pass

try:
    links_nomes = [x.get_text() for x in tags_a]
except AttributeError:
    pass

print(links)
```

```python
print(links_nomes)
```

```python
pre_autor = [x.get_text(separator='\n') for x in soup.find_all('li')]
autor_uni = [i.split('\n') for i in pre_autor]
autores_e_enti = []

for i in range(0, len(autor_uni)):
    try:
        autores_e_enti.append(autor_uni[i][1])
    except IndexError:
        pass
```

```python
print(len(autor_uni))
print(len(autores_e_enti))
```

```python
data = pd.DataFrame(
    {
        'títulos_2020': links_nomes,
        'links_2020': links,
        'autores_e_universidades_2020': autores_e_enti
    }
)
print(data)
```

```python
data = data[data['links_2020'].str.contains(".docx?$|.pdf$")]
```

```python
data.loc[:, "code_area"] = data['links_2020'].str.extract(r"(?P<area>\/i[0-9]{1,2})")['area'].str.replace(r"\/", "")
print(data.head())
```

```python
areas_df = pd.DataFrame(
    {
        'code_area': [f'i{n}' for n in range(1, 14)],
        'area': areas
    }
)
print(areas_df)
```

```python
data_final = pd.merge(left=data, right=areas_df, on='code_area', how='outer')
print(data_final.links_2020.head())
```
